```python
#Lambda for splittting
Merge_DF_all_plates_2p5_col[['CellLine Name', "Plate"]] = Merge_DF_all_plates_2p5_col["Customer Order Number"].apply(lambda x: pd.Series(str(x).split("_2.5_")))
#Lambda for mergerging two columns
Merge_DF_all_plates_2p5_col['Pos_CellLine'] = Merge_DF_all_plates_2p5_col.apply(lambda row: str(row['position']) + ';' + row['CellLine Name'], axis=1)
```

```python
# Convert columns to numeric, coercing errors to NaN
Merge_DF_all_plates_2p5_col['reads_total'] = pd.to_numeric(Merge_DF_all_plates_2p5_col['reads_total'], errors='coerce')
Merge_DF_all_plates_2p5_col['af'] = pd.to_numeric(Merge_DF_all_plates_2p5_col['af'], errors='coerce')

# Handle missing values (if any)
Merge_DF_all_plates_2p5_col['reads_total'].fillna(0, inplace=True)
Merge_DF_all_plates_2p5_col['af'].fillna(0, inplace=True)

#Groupby using lambda function
result = Merge_DF_all_plates_2p5_col.groupby('Pos_CellLine').agg({'reads_total': [lambda x: x.mean(), lambda x: x.std()], 'af': [lambda x: x.mean(), lambda x: x.std()]}).reset_index()
result.columns = ['Pos_CellLine', 'reads_total_mean', 'reads_total_std', 'af_mean', 'af_std']
```
