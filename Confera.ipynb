```python
df6 = pd.DataFrame(df4, columns=['SAN_ccmf_perfect',"CAM_ccmf_perfect"]) #,'SAN_GC_Dropout','CAM_GC_Dropout'])
df6.head()
```

```python
vals, names, xs = [],[],[]
for i, col in enumerate(df6.columns):
    vals.append(df6[col].values)
    names.append(col)
    xs.append(np.random.normal(i + 1, 0.04, df6[col].values.shape[0]))  # adds jitter to the data points - can be adjusted
```

```python
plt.figure (figsize=(10,10))
sns.set_style("whitegrid")
palette = ['#33FF3B', '#3379FF', '#FFD633', '#33FFF1']
plt.title("ccmf_perfect- Healthy Samples", fontsize=18)
plt.xlabel("Samples", fontweight='normal', fontsize=14)
plt.ylabel("", fontweight='normal', fontsize=14)
plt.axhline(3)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

#sns.despine(bottom=True) # removes right and top axis lines
#plt.axhline(y=15, color='#ff3300', linestyle='--', linewidth=1, label='Threshold Value = 5')
#plt.legend(bbox_to_anchor=(0.31, 1.06), loc=2, borderaxespad=1., framealpha=1, facecolor ='white', frameon=True)


plt.boxplot(vals, labels=names, notch=False)#, #, boxprops=boxprops, whiskerprops=whiskerprops,capprops=capprops, flierprops=flierprops, medianprops=medianprops,showmeans=False)
palette = ['r', 'g', 'b', 'y']
for x, val, c in zip(xs, vals, palette):
    plt.scatter(x, val, alpha=0.4, color=c)
plt.show()
```

```python
plt.figure (figsize=(10,10))
sns.set_style("whitegrid")
palette = ['#33FF3B', '#3379FF', '#FFD633', '#33FFF1']
plt.title("Correlation San Diego/Cambridge Labs", fontsize=18)
plt.xlabel("CAM_ccmf_perfect", fontweight='normal', fontsize=14)
plt.ylabel("SAN_ccmf_perfect", fontweight='normal', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
X = df6 ['SAN_ccmf_perfect']
Y1 = df6['CAM_ccmf_perfect']
#Y2 = X**2 + np.random.random(X.shape)

plt.scatter(X,Y1,color='g')
#plt.scatter(X,Y2,color='g')
plt.plot(np.unique(X), np.poly1d(np.polyfit(X, Y1, 0.1))(np.unique(X)), color='yellow')
plt.show()
```
